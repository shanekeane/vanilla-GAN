```python
from vanilla_gan import train, utils
```

# Set parameters

```python
BATCH_SIZE=128
NOISE_DIM=100
DATA_TYPE="mnist"
SAMPLES_PLOT=6
EPOCHS=50
```

# Load data

```python
data_ldr = utils.load_data(DATA_TYPE, BATCH_SIZE);
```

# Train and plot Vanilla GAN with MNIST data

```python
generator, gen_loss, disc_loss = train.train_vanilla_gan(dataloader=data_ldr,
                                                   epochs=EPOCHS, 
                                                   samples_to_plot=SAMPLES_PLOT, 
                                                   batch_sz=BATCH_SIZE,
                                                   noise_dim=NOISE_DIM)
```

# Plot some samples

```python
utils.plot_samples(generator, no_samples=10, noise_dim=NOISE_DIM)
```
